```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from plot import Plots
```

```python
plt.style.use('fivethirtyeight')

path = "https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json"
df = pd.read_json(path)
```

### Model statistic

```python
def accuracy_score(y_true, y_pred):
    return sum(y_true == y_pred) / len(y_true)

def rmse(y_true, y_pred):
    mean = np.square(y_true - y_pred).mean()
    return np.sqrt(mean)
```

```python
accuracy = accuracy_score(df.gt_corners, df.rb_corners)
print("Model accuracy:", accuracy)

# Calculate RMSE with floor_mean and ceiling_mean
rmse_score = rmse(df.floor_mean, df.ceiling_mean)
print("RMSE with floor_mean and ceiling_mean:", rmse_score)
```
